**Descriptive Analytics** - a statistical method that is used to *search* and *summarize* historical data in order to identify patterns or meaning.<br>
> - Answers the question "What happened?"<br>
> - Meant to provide insight into historical patterns

<img src='Week4_fig1.png'/>

> - Helps a business understand how it is performing by providing context to help stakeholders _interpret information_<br>
> - Explore raw data from multiple sources to give valuable insights<br>
> - Signal that something has happened which maybe right or wrong without explaining why

#### **Examples**

* Summarizing historical events such as geographical sales, customer churn, performance of marketing campaigns, etc.
* Tabulation of social metrics such as Facebook likes, Tweets or subscribers
* Tracking course enrollments and course compliance rates in a learning management system (LMS)

#### **2 Parts of Descriptive Analytics**<br>
1. **Exploratory Data Analysis** - involves preliminary exploration of data set prior to a more rigorous analyses<br>
> * Data profiling
> * Univariate and bvariate statistics
> * Associations<br>
2. **Focused Descriptive Analytics** - answering specific business questions<br>
> * What were the sales last quarter?
> * What was the least sold product last year?
