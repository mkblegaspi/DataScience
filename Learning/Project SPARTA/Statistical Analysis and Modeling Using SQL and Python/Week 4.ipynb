 **Descriptive Analytics** - a statistical method that is used to *search* and *summarize* historical data in order to identify patterns or meaning.<br>
- Answers the question "What happened?"<br>
- Meant to provide insight into historical patterns

- Helps a business understand how it is performing by providing context to help stakeholders _interpret information_<br>
- Explore raw data from multiple sources to give valuable insights<br>
- Signal that something has happened which maybe right or wrong without explaining why

**Examples**
* Summarizing historical events such as geographical sales, customer churn, performance of marketing campaigns, etc.
* Tabulation of social metrics such as Facebook likes, Tweets or subscribers
* Tracking course enrollments and course compliance rates in a learning management system (LMS)

>>* Associations<br>
>2. **Focused Descriptive Analytics** - answering specific business questions<br>

### **4.2.2 Diagnostic Analytics**

* Answers the question "Why did it happen?"
* Examines data and content to answer the "Why".
* Powerful diagnostic analytical tools will empower the analyst to drill down data and potentialy isolate of the root cause of the problem

**Diagnostic Analytics Functions**
* Identify Anomalies
* Drill Down (Data Discovery)
* Determine Causal Relationships

>**Identification of Anomalies**
>* identification of areas that require further investigation
>* more questions may be asked from the previous phase
>* Examples:
    * Why did the revennues increased on this certain region?
    * Why is there a sudden build upp of hits in the company website?
    * Why did a lot of customer double their orders on this particular product?
    * Which department caused the performance of the whole company to plunge down?

> **Drill Down (Data Discovery)**
>* identification of data sources that will help explain the anomalies
>* pattern discovery, checking of data outside the organization, external events

> **Determining Causal Relationships**
> * Hidden relationships may be uncovered looking at external events
> * Correlations, data filtering and further data drill down

**Diagnostic Analytics Examples**
* Drilling down data according to geography to identify its different effects on sales
* Identifying the correlation between external/world events on company performance
* Isolating products that contibute to negative revenues in reports

**Diagnostic Analytics Tools**
* Software capable of drill down analytics
* Measures od association
* Regression analysis
* Time series Analysis
* Data Filtering
